```python
import os
import shutil

# File paths
base_path = "/content/drive/MyDrive/PAN19/PAN/Train/en/"  # Replace with your folder path
txt_file = os.path.join(base_path, "truth.txt")  # Replace with your txt file name

# Create subfolders
folders = ["bot", "male", "female"]
for folder in folders:
    os.makedirs(os.path.join(base_path, folder), exist_ok=True)

# Categorize IDs from the txt file
categories = {"bot": [], "female": [], "male": []}

# Read and categorize entries from the txt file
with open(txt_file, "r") as file:
    for line in file:
        parts = line.strip().split(":::")
        if len(parts) == 3:
            identifier, category, subcategory = parts
            if category == "bot":
                categories["bot"].append(identifier)
            elif category == "human":
                if subcategory == "female":
                    categories["female"].append(identifier)
                elif subcategory == "male":
                    categories["male"].append(identifier)

# Move XML files to their respective subfolders
for filename in os.listdir(base_path):
    if filename.endswith(".xml"):
        file_id = os.path.splitext(filename)[0]  # Extract ID from the XML file name
        for category, ids in categories.items():
            if file_id in ids:
                # Move the file to the corresponding folder
                source = os.path.join(base_path, filename)
                destination = os.path.join(base_path, category, filename)
                shutil.move(source, destination)
                break

print("XML files have been organized into subfolders!")
```

```python
import os
import csv
from xml.etree import ElementTree as ET

# Specify the folder containing the XML files
xml_folder = "/content/drive/MyDrive/PAN19/PAN/Train/en/male/"  # Replace with your XML folder path
output_folder = "/content/drive/MyDrive/PAN19/PAN/Train/en/"  # Output CSV folder
os.makedirs(output_folder, exist_ok=True)  # Ensure output folder exists

output_csv = os.path.join(output_folder, "male_tweets.csv")

# Create a CSV file
with open(output_csv, mode="w", newline="", encoding="utf-8") as csv_file:
    csv_writer = csv.writer(csv_file)
    # Write header
    csv_writer.writerow(["author_id", "lang", "text"])

    # Process each XML file in the folder
    for file_name in os.listdir(xml_folder):
        if file_name.endswith(".xml"):
            file_path = os.path.join(xml_folder, file_name)
            try:
                # Parse the XML file
                tree = ET.parse(file_path)
                root = tree.getroot()

                # Extract author language
                lang = root.get("lang", "unknown")

                # Extract all <document> contents
                for document in root.findall(".//document"):
                    # Handle CDATA sections and strip whitespace
                    text = document.text.strip() if document.text else ""
                    # Write author_id, lang, and text to the CSV
                    csv_writer.writerow([file_name, lang, text])
            except ET.ParseError as e:
                print(f"Error parsing {file_name}: {e}")
            except Exception as e:
                print(f"Unexpected error with {file_name}: {e}")

print(f"CSV file has been created: {output_csv}")
```

```python
import os
import pandas as pd

# Define file paths for your CSVs
csv_files = [
    "val_male_tweets.csv",
    "val_female_tweets.csv",
    "val_bot_tweets.csv",
    "train_male_tweets.csv",
    "train_female_tweets.csv",
    "train_bot_tweets.csv",
    "test_male_tweets.csv",
    "test_female_tweets.csv",
    "test_bot_tweets.csv"
]

# Folder where CSVs are stored
input_folder = "/content/drive/MyDrive/PAN19/PAN/Cleaned CSVs"
output_file = os.path.join(input_folder, "merged_tweets.csv")

# Function to determine the category (male/female/bot) based on file name
def get_category(file_name):
    if "male" in file_name:
        return "male"
    elif "female" in file_name:
        return "female"
    elif "bot" in file_name:
        return "bot"
    return "unknown"

# Function to determine the split (train/val/test) based on file name
def get_split(file_name):
    if "train" in file_name:
        return "train"
    elif "val" in file_name:
        return "val"
    elif "test" in file_name:
        return "test"
    return "unknown"

# Initialize an empty DataFrame to store merged data
merged_data = pd.DataFrame()

# Process each CSV file
for file_name in csv_files:
    file_path = os.path.join(input_folder, file_name)
    if os.path.exists(file_path):  # Check if the file exists
        # Load the CSV into a DataFrame
        data = pd.read_csv(file_path)
        # Add new columns: 'category' and 'split'
        data['category'] = get_category(file_name)
        data['split'] = get_split(file_name)
        # Append the current data to the merged DataFrame
        merged_data = pd.concat([merged_data, data], ignore_index=True)
    else:
        print(f"File not found: {file_name}")

# Save the merged DataFrame to the output CSV
merged_data.to_csv(output_file, index=False, encoding="utf-8")

print(f"Merged data saved to {output_file}")
```

```python
import pandas as pd

def clean_dataframe(df):
    df = df.sample(frac=1, random_state=42)
    df = df.drop_duplicates()
    df = df.dropna()
    return df

# Clean all datasets
train_df = clean_dataframe(train_df)
test_df = clean_dataframe(test_df)
val_df = clean_dataframe(val_df)

# Save cleaned datasets to CSV
train_df.to_csv("/content/drive/MyDrive/PAN19/PAN/train_tweets_cleaned.csv", index=False)
test_df.to_csv("/content/drive/MyDrive/PAN19/PAN/test_tweets_cleaned.csv", index=False)
val_df.to_csv("/content/drive/MyDrive/PAN19/PAN/val_tweets_cleaned.csv", index=False)
```
